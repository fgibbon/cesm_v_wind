```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
```

### Loading in data from CEDA

```python
ds_ap = xr.open_dataset('/badc/cmip6/data/CMIP6/CFMIP/NCAR/CESM2/aqua-control/r1i1p1f1/Amon/va/gn/latest/va_Amon_CESM2_aqua-control_r1i1p1f1_gn_000101-001012.nc')
ds_amip = xr.open_dataset('/badc/cmip6/data/CMIP6/CMIP/NCAR/CESM2/amip/r1i1p1f1/Amon/va/gn/latest/va_Amon_CESM2_amip_r1i1p1f1_gn_195001-201412.nc')
ds_hist = xr.open_dataset('/badc/cmip6/data/CMIP6/CMIP/NCAR/CESM2/historical/r1i1p1f1/Amon/va/gn/latest/va_Amon_CESM2_historical_r1i1p1f1_gn_185001-201412.nc')
ds_pic = xr.open_dataset('/badc/cmip6/data/CMIP6/CMIP/NCAR/CESM2/piControl/r1i1p1f1/Amon/va/gn/latest/va_Amon_CESM2_piControl_r1i1p1f1_gn_000101-009912.nc')

datasets = [ds_amip, ds_hist, ds_pic, ds_ap]
labels = ['AMIP', 'Historical', 'piControl', 'Aquaplanet']
```

### Plotting V

The pattern is stronger in the time mean but I've included the snapshots to show it's present in the raw data.

Changing the range of latitudes plotted can make the pattern easier to see

For the residual plot: we smooth the data over 4 latitude coordinates and subtract the smoothed data from the raw data, leaving a residual which shows the pattern more clearly.

```python
# Snapshot in time

fig, axs = plt.subplots(
    nrows=4,
    ncols=3,
    figsize=(15, 20)
)
fig.subplots_adjust(hspace=0.5)  

for i, (ds,label) in enumerate(zip(datasets,labels)):
    V250 = ds['va'].sel(plev=25000) # Can change the height
    V250_t = V250.isel(time=60) # Can change the time
    V250_t = V250_t.sel(lat=slice(-90,-40)) # Just looking at high latitude region to see pattern more clearly / can comment out

    # Plotting maps
    V250_t.plot(ax=axs[i, 0])
    axs[i, 0].set_title('V @ 250 hPa time snapshot')
    
    # Zonal means
    zm = V250_t.mean('lon')
    axs[i, 1].plot(zm, zm.lat)
    axs[i, 1].set_title('Zonal Mean')
    
    # Removing smoothed V field
    smoothV_t = V250_t.rolling(lat=4, center=True).mean()
    v_t = V250_t - smoothV_t
    v_t.plot(ax=axs[i, 2])
    axs[i, 2].set_title('Residual Artefact')

    y = axs[i, 0].get_position().y1 + 0.02
    fig.text(0.5, y, label, ha='center', va='bottom', fontsize=14)

    
    
```

```python
# Time mean

fig, axs = plt.subplots(
    nrows=4,
    ncols=3,
    figsize=(15, 20)
)
fig.subplots_adjust(hspace=0.5)  

for i, (ds,label) in enumerate(zip(datasets,labels)):
    V250 = ds['va'].sel(plev=25000) # Can change the height
    V250_tm = V250.mean('time') # Can change the time
    V250_tm = V250_tm.sel(lat=slice(-90,-40)) # Just looking at high latitude region to see pattern more clearly / can comment out

    # Plotting maps
    V250_tm.plot(ax=axs[i, 0])
    axs[i, 0].set_title('V @ 250 hPa time mean')
    
    # Zonal means
    zm = V250_tm.mean('lon')
    axs[i,1].plot(zm, zm.lat)
    axs[i, 1].set_title('Zonal Mean')
    
    # Removing smoothed V field
    smoothV_tm = V250_tm.rolling(lat=4, center=True).mean()
    v_tm = V250_tm - smoothV_tm
    v_tm.plot(ax=axs[i, 2])
    axs[i, 2].set_title('Residual Artefact')

    y = axs[i, 0].get_position().y1 + 0.02
    fig.text(0.5, y, label, ha='center', va='bottom', fontsize=14)

    
    
```
